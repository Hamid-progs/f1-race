```python
def parse_time(x):
    x = str(x)
    
    # Numeric race time
    if ':' in x:
        parts = x.split(':')
        if len(parts) == 3:  # h:m:s
            h, m, s = parts
        elif len(parts) == 2:  # m:s
            h = 0
            m, s = parts
        s = float(s)
        return int(h)*3600 + int(m)*60 + s
    
    # Time behind leader
    elif x.startswith('+'):
        if 'lap' in x:
            laps = int(x.split()[0][1:])  # '+2 laps' → 2
            # Option: approximate each lap as 90 seconds (depends on track)
            return laps * 90
        elif 's' in x:
            seconds = float(x[1:-1])
            return seconds
        else:
            return None
    
    # Retirements or special codes
    elif x in ['Ret','DNF','DNS','DSQ','Acc','Mechanical','NC','DQ']:
        return np.nan
    
    else:
        return np.nan
```

```python
df['Time_seconds'] = df['Time / Retired'].apply(parse_time)
```
