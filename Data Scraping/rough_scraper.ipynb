```python
import pandas as pd 
import numpy as np 

import re
import requests
from bs4 import BeautifulSoup as bs
```

```python
# apply a for loop to loop the years
initial_link = 'https://www.formula1.com/en/results/2015/races'
```

```python
r = requests.get('https://www.formula1.com/en/results/2015/races')
web_page = bs(r.content)
print(web_page.prettify())
```

```python
h1 = web_page.find("h1")
year = re.search(r"\b(201[5-9]|202[0-5])\b", h1.get_text()).group()

print(year)  # 2015
```

```python
tds = web_page.find_all("td",class_="typography-module_body-s-semibold__O2lOH Table-module_cell__3rpTC Table-module_no-wrap-text__CP2oI")
dates = []
for td in tds:
    date = td.get_text(strip=True)
    dates.append(date)
    print(date)  # Australia
dates
```

```python
dates = []

for row in web_page.select("tr.Table-module_body-row__shKd-"):
    tds = row.find_all("td")
    if len(tds) > 1:
        date = tds[1].get_text(strip=True)
        dates.append(date)
        print(date)  # 15 Mar
```

```python
a_tag = web_page.find_all("a", class_="flex gap-px-10 items-center")
```

```python
countries = []
for tag in a_tag:
    country = tag.find(text=True, recursive=False).strip()
    countries.append(country)
    print(country)  # Australia
countries
```
